```python
import librosa
#extract MFCCs from audio file
def extract_features(file):
  x, sr = librosa.load(os.path.join(file), res_type = "kaiser_fast") #load the audio file into librosa
  mfccs = (librosa.feature.mfcc(y = x, sr = sr, n_mfcc = 40))
  return mfccs
```

```python
#create dict to represent the emotion labels from the ravdess dataset
emotions = {"01": "neutral",
            "02": "calm",
            "03": "happy",
            "04": "sad",
            "05": "angry",
            "06": "fearful",
            "07": "disgust",
            "08": "surprised"}
emotion_map = {"happy": 1, "angry": 2, "calm": 3, "sad": 4} #use numerical values in data for better training
observed_emotions = list(emotions.values())
```

```python
import glob,os
import librosa
x, y = [], [] #lists for x and y
#path = "drive/MyDrive/archive/Actor_*/*.wav"

path = "drive/MyDrive/archive/Actor_*/*.wav"

for file in glob.glob(path):
  filename = os.path.basename(file)

  #gets the third number from filename and then selects the same key from the emotions dict
  emotion = emotions[filename.split("-")[2]]
  #choose songs that have an emotion of happy, calm, angry, or sad
  if emotion not in observed_emotions or emotion in ["neutral", "fearful", "disgust", "surprised"]:
    continue
  feature = extract_features(file)
  x.append(feature)
  y.append(emotion_map[emotion])

print(len(x),len(y))
```

```python
import numpy as np

# explicit function to normalize array
def normalize(arr, t_min, t_max):
    norm_arr = []
    diff = t_max - t_min
    diff_arr = max(arr) - min(arr)
    for i in arr:
        temp = (((i - min(arr))*diff)/diff_arr) + t_min
        norm_arr.append(temp)
    return norm_arr

#resize MFCC arrays to be standard size
x = [np.resize(i,(40,127)) for i in x]
```
